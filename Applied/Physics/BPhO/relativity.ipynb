```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)

from qbstyles import mpl_style

mpl_style(dark=False)
```

Consider a particle, with initial speed $u$, at the top of a rectangular box with height $L$. Assume there is no gravity. The time, $\Delta t$ it takes for the particle to fall a distance $L$ is

$$
\Delta t = \frac{L}{u}.
$$

An observer watches the box, along with the particle, travel horizontally at a constant speed $v$. The observer watches the particle fall (from $t=0$ to $t=\Delta t$), covering a vertical distance of $L$ and a horizontal distance of $v\Delta t$. From the observer's perspective, the particle moves along a diagonal line of length $d$.

$$
d = \sqrt{L^2 + v^2 \Delta t^2} = \sqrt{L^2 + \frac{v^2 L^2}{u^2}} = L \sqrt{1 + \frac{v^2}{u^2}}
$$

The speed of the particle, $w$, is

$$
w = \frac{d}{\Delta t} = d \frac{u}{L} = u \sqrt{1 + \frac{v^2}{u^2}}
$$

If this were a particle were a pulse of light, it would appear to be the case that the speed of light is different in different reference reference frames. However, this is not correct. The speed of light is always c regardless of the reference frame.

$$
c = \frac{1}{\sqrt{\mu_0 \varepsilon_0}} = 2.998 \times 10^8 \text{ ms}^{-1}
$$

Instead, let time progress at a different rate depending on the relative motion of two frames of reference. Let $t'$ denote the time that passes in the box frame of reference, wheareas $t$ is the time that passes in the observer's frame of reference. When the light pulse hits the bottom of the box, $t' = \Delta t' = L/c$. Now, $w$, which must be equal to $c$, is

$$
\begin{align*}
w &= \frac{d}{\Delta t} = \frac{\sqrt{L^2 + v^2\Delta t^2}}{\Delta t} \\
c^2 \Delta t^2 &= L^2 + v^2 \Delta t^2 \\
L &= c \Delta t \sqrt{1 - \frac{v^2}{c^2}} \\
\Delta t' &= \Delta t \sqrt{1 - \frac{v^2}{c^2}}
\end{align*}
$$

The time experienced in the box, $\Delta t'$ can be related to the time experienced by the observer, $\Delta t$ with

$$
\Delta t' = \frac{\Delta t}{\gamma}
$$

where $\gamma$, the Lorentz factor, is

$$
\gamma = \left(1 - \frac{v^2}{c^2}\right)^{-\frac{1}{2}}
$$

This is time dilation. With respect to the time passing in some reference frame, time will pass slower in another reference frame which moves. When $v \ll c$, $\gamma \approx 1$. Thus, $\Delta t' \approx \Delta t$.

```python
v_over_c = np.linspace(0, 0.999, 5000)

gamma = (1 - v_over_c ** 2) ** (-1/2)
```

```python
plt.plot(v_over_c, gamma)

plt.xlim(0, 1)
plt.ylim(0, 10)

plt.xlabel(r"$v/c$")
plt.ylabel(r"Lorentz Factor, $\gamma$")

plt.title(r"$\gamma$ vs $v/c$")

plt.show()
```
