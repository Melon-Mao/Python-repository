The Atwood machine was invented in 1784 by George Atwood as a laboratory experiment to verify the mechanical laws of motion with constant acceleration. The Atwood machine consists of two masses, $m_1$ and $m_2$, connected by a string that passes over a pulley. The pulley is assumed to be massless and frictionless. The string is assumed to be massless and inextensible (i.e., its length is constant). The masses are assumed to be point masses. The system is assumed to be in a uniform gravitational field, with the positive $y$-direction pointing upward. The acceleration of gravity is $g$.

If we wish to find the equations of motion for the Atwood machine via Newton's laws of motion, we will use his second law:

$$
\sum \vec{F} = m \vec{a}
$$

where $\sum \vec{F}$ is the sum of all forces acting on the system, $m$ is the total mass of the system, and $\vec{a}$ is the acceleration of the system. While we could get tw0 equations of motion by applying Newton's second law to each mass, we can describe the entire system in terms of the displacement $x$, which is the distance that $m_1$ has moved downward from its initial position. The displacement of the second mass is then $l-x$, where $l$ is the length of the string. The acceleration of the system is then $\ddot{x}$, which is the second derivative of $x$ with respect to time. The equations of motion are then:

$$
\sum F = m \ddot{x}
$$

where $\sum F$ is the sum of all forces acting on the system. The forces acting on the system are the gravitational force on each mass and the tension in the string. The gravitational force on each mass is given by:

$$
\begin{align}
F_{g1} &= m_1 g \\
F_{g2} &= -m_2 g
\end{align}
$$

This is because the gravity vector is always pointing downward, in the same direction of the displacement $x$ for $m_1$ and in the opposite direction of the displacement $l-x$ for $m_2$. The tension in the string is the same for both masses and may be denoted by $T$. If $\ddot x$ is positive when $m_1$ is accelerating downward, then the force of gravity on $m_1$ is positive and the tension force is negative. The inverse is true for $m_2$. The equations for acceleration are then:

$$
\begin{align}
m_1 g - T &= m_1 \ddot{x} \\
T - m_2 g &= m_2 \ddot{x}
\end{align}
$$

We add the two equations together to eliminate $T$ and obtain:

$$
\begin{align}
m_1 g - m_2 g &= (m_1 + m_2) \ddot{x} \\
\ddot{x} &= g \frac{m_1 - m_2}{m_1 + m_2}
\end{align}
$$

Now, we may integrate the equation for acceleration twice to obtain the equation for displacement:

$$
\begin{align}
\int \ddot{x} dt &= \int g \frac{m_1 - m_2}{m_1 + m_2} dt \\
\dot{x} &= g \frac{m_1 - m_2}{m_1 + m_2} t + C_1 \\
\int \dot{x} dt &= \int g \frac{m_1 - m_2}{m_1 + m_2} t + C_1 dt \\
x &= \frac{1}{2} g \frac{m_1 - m_2}{m_1 + m_2} t^2 + C_1 t + C_2
\end{align}
$$

By solving for $x(0)$ and $\dot{x}(0)$, we find that the constants of integration represent the initial position and velocity of the first mass. This is because the displacement function $x(t)$ is the displacement of the first mass. We may then write the equations of motion as:

$$
\begin{align}
x(t) &= \frac{1}{2} g \frac{m_1 - m_2}{m_1 + m_2} t^2 + x_0 + \dot{x}_0 t \\
\dot{x}(t) &= g \frac{m_1 - m_2}{m_1 + m_2} t + \dot{x}_0
\end{align}
$$

where $x_0$ and $\dot{x}_0$ are the initial position and velocity of the first mass, respectively. We may also write the equations of motion in terms of the displacement of the second mass, $l-x$, by substituting $l-x$ for $x$ in the equations above. The equations of motion are then:

$$
\begin{align}
l-x(t) &= \frac{1}{2} g \frac{m_1 - m_2}{m_1 + m_2} t^2 + x_0 + \dot{x}_0 t \\
-\dot{x}(t) &= g \frac{m_1 - m_2}{m_1 + m_2} t + \dot{x}_0
\end{align}
$$

where $x_0$ and $\dot{x}_0$ are the initial position and velocity of the second mass, respectively.

We have looked at the Newtonian approach of obtaining the equations of motion for the Atwood machine. We may also obtain the equations of motion using the Lagrangian approach. This looks at the energies of the system instead of the forces. We can find the equations of motion, first by finding the Lagrangian of the system, then by finding the Euler-Lagrange equations. The Lagrangian of the system is:

$$
\begin{align}
\mathcal{L} &= T - V \\
T &= \frac{1}{2} m_1 v_1^2 + \frac{1}{2} m_2 v_2^2 \\
V &= m_1 g h_1 + m_2 g h_2
\end{align}
$$

where $T$ is the kinetic energy of the system and $V$ is the potential energy. We know that $v_1$ and $v_2$ are the velocities of $m_1$ and $m_2$, respectively, and that $v_1 = \dot{x}$, $v_2 = -\dot{x}$. Also, $h_1$ and $h_2$ are the heights of $m_1$ and $m_2$. That means that $h_1 = -x$ and $h_2 = -(l-x)$. We can now substitute these values into the Lagrangian to obtain:

$$
\begin{align}
\mathcal{L} &= \frac{1}{2} m_1 \dot{x}^2 + \frac{1}{2} m_2 \dot{x}^2 + m_1 g x - m_2 g (l+x) \\
&= \frac{1}{2} (m_1 + m_2) \dot{x}^2 + m_1 g x - m_2 g x - m_2 g l
\end{align}
$$

We can know use the Euler-Lagrange equations to find the equations of motion. The Euler-Lagrange equations are:

$$
\frac{d}{dt} \left( \frac{\partial \mathcal{L}}{\partial \dot{x}} \right) - \frac{\partial \mathcal{L}}{\partial x} = 0
$$

To get the partial derivative of the Largrangian in respect to $\dot{x}$, we treat everything else as a constant whilst differentiating. This gives us:

$$
\frac{\partial \mathcal{L}}{\partial \dot{x}} = (m_1 + m_2) \dot{x}
$$

We may also get the partial derivative of the Lagrangian in respect to $x$ by treating everything else as a constant whilst differentiating. This gives us:

$$
\frac{\partial \mathcal{L}}{\partial x} = m_1 g - m_2 g
$$

The rest of the equation is as follows:

$$
\begin{align}
\frac{d}{dt} \left( \frac{\partial \mathcal{L}}{\partial \dot{x}} \right) - \frac{\partial \mathcal{L}}{\partial x} &= 0 \\
&= \frac{d}{dt} \left( (m_1 + m_2) \dot{x} \right) - (m_1 g - m_2 g) \\
&= (m_1 + m_2) \ddot{x} - (m_1 g - m_2 g) \\
&= 0
\end{align}
$$

Solving for $\ddot{x}$ gives us a familiar equation:

$$
\begin{align}
(m_1 + m_2) \ddot{x} - (m_1 g - m_2 g) &= 0 \\
\ddot{x} &= g \frac{m_1 - m_2}{m_1 + m_2}
\end{align}
$$

We have obtained the equations of motion, but both the Lagrangian and Newtonian methods haven't gotten as the true equations. This is not a problem with the methods, but rather because we have forgetten one important thing. That is the rotational inertia, which shows how much a body resists rotational motion. This is more important for small differences in mass between $m_1$ and $m_2$. The no-slip condition gives us a formula for angular acceleration, $\alpha$, in terms of linear acceleration, $\ddot{x}$:

$$
\alpha = \frac{\ddot{x}}{r}
$$

where $r$ is the radius of the pulley. We can easily add this to our equations of motion with the Lagrangian approach, as this can be summed with the kinetic energy. We will need to get the angular velocity, $\omega$, in terms of $\dot{x}$. We know that $\omega = \frac{\dot{x}}{r}$, and that the rotational inertia times angular velocity squared is the rotational kinetic energy. This gives us:

$$
\begin{align}
\mathcal{L} &= T - V \\
T &= \frac{1}{2} m_1 v_1^2 + \frac{1}{2} m_2 v_2^2 + \frac{1}{2} I \frac{\dot{x}^2}{r^2} \\
V &= m_1 g h_1 + m_2 g h_2
\end{align}
$$

Now let's do the rest of the steps:

$$
\begin{align}
\mathcal{L} &= \frac{1}{2} m_1 \dot{x}^2 + \frac{1}{2} m_2 \dot{x}^2 + \frac{1}{2} I \frac{\dot{x}^2}{r^2} + m_1 g x - m_2 g (l+x) \\
&= \frac{1}{2} (m_1 + m_2) \dot{x}^2 + \frac{1}{2} I \frac{\dot{x}^2}{r^2} + m_1 g x - m_2 g x - m_2 g l
\end{align}
$$

$$
\begin{align}
\frac{\partial \mathcal{L}}{\partial \dot{x}} &= (m_1 + m_2) \dot{x} + \frac{I \dot{x}}{r^2} \\
\frac{\partial \mathcal{L}}{\partial x} &= m_1 g - m_2 g
\end{align}
$$

$$
\begin{align}
\frac{d}{dt} \left( \frac{\partial \mathcal{L}}{\partial \dot{x}} \right) - \frac{\partial \mathcal{L}}{\partial x} &= 0 \\
&= \frac{d}{dt} \left( (m_1 + m_2) \dot{x} + \frac{I \dot{x}}{r^2} \right) - (m_1 g - m_2 g) \\
&= (m_1 + m_2) \ddot{x} + \frac{I \ddot{x}}{r^2} - (m_1 g - m_2 g) \\
&= 0
\end{align}
$$

$$
\begin{align}
(m_1 + m_2) \ddot{x} + \frac{I \ddot{x}}{r^2} - (m_1 g - m_2 g) &= 0 \\
\ddot{x} &= g \frac{m_1 - m_2}{m_1 + m_2 + \frac{I}{r^2}}
\end{align}
$$

```python
# Create symbolic variables
m_1, m_2, ell, g, t = sp.symbols('m_1, m_2, ell, g, t')
I, r = sp.symbols('I, r')
gamma = sp.symbols('gamma')
x = dynamicsymbols('x')

# Initial conditions
x_0 = sp.symbols('x_0')
v_0 = sp.symbols('v_0')

# Derivative of x
x_dot = sp.diff(x, t)
x_ddot = sp.diff(x_dot, t)

# Lagrangian
T_1 = sp.Rational(1, 2)*m_1*x_dot**2
T_2 = sp.Rational(1, 2)*m_2*(-x_dot**2) # v_1 = -v_2
T_3 = sp.Rational(1, 2)*I*x_dot**2/r**2

T = T_1 + T_2 + T_3

V_1 = -m_1*g*x
V_2 = m_2*g*(ell + x)

V = V_1 + V_2


L = T - V
# L *= sp.E ** (-gamma*t)
```

```python
print('Lagrangian:')
L.simplify()
```

### Euler-Lagrange Equations

$$
\dfrac{d}{dt} \left( \dfrac{\partial \mathcal{L}}{\partial \dot q_i} \right) - \dfrac{\partial \mathcal{L}}{\partial q_i} = 0
$$

$$
\big \Downarrow
$$

$$
\frac{d}{dt} \left( \frac{\partial \mathcal{L}}{\partial \dot{x}} \right) - \frac{\partial \mathcal{L}}{\partial x} = 0
$$

```python
eq = sp.Eq(sp.diff(sp.diff(L, x_dot), t) - sp.diff(L, x), 0).simplify()

eq
```

```python
solution = sp.solve(eq, x_ddot)[0].simplify()

sp.Eq(x_ddot, solution)
```

```python
sol = sp.dsolve(eq, x).rhs

sol
```

```python
sol_dot = sp.diff(sol, t)

sol_dot
```

```python
print("Initial position:")
sol.subs(t, 0)
```

```python
print("Initial velocity:")
sol_dot.subs(t, 0)
```

```python
sol = sol.replace(sp.symbols("C1"), x_0).replace(sp.symbols("C2"), v_0)
sol_dot = sol_dot.replace(sp.symbols("C1"), x_0).replace(sp.symbols("C2"), v_0)
```

```python
sol
```

```python
sol_dot
```

```python
pos_1 = sol
pos_2 = ell - sol

vel_1 = sol_dot
vel_2 = -sol_dot
```

```python
# Initial conditions
param_g = 9.81 # ms^-2 (gravitational acceleration)

param_m_1 = 2.0 # kg (mass of the first block)
param_m_2 = 1.5 # kg (mass of the second block)
param_m_pulley = 0 # kg (mass of the pulley)

param_gamma = 3 # s^-1 (damping coefficient)

param_ell = 15.0 # m (length of the rope)

param_r = 3 # m (radius of the pulley)
param_I = 0.5*param_m_pulley*param_r**2 # kg m^2 (moment of inertia of the pulley)

param_x_0 = 2.0 # m (initial position of the first block)
param_v_0 = 0.0 # ms^-1 (initial velocity of the first block)

parameters = (param_g, param_m_1, param_m_2, param_ell, param_x_0, param_v_0, param_I, param_r, param_gamma)

# Time array
t_0 = 0
t_f = 10
time = np.linspace(t_0, t_f, (t_f - t_0) * 25) # @ 25 fps

# Symbolic to numeric functions
pos_1_f = sp.lambdify((t, g, m_1, m_2, ell, x_0, v_0, I, r, gamma), pos_1)
pos_2_f = sp.lambdify((t, g, m_1, m_2, ell, x_0, v_0, I, r, gamma), pos_2)

vel_1_f = sp.lambdify((t, g, m_1, m_2, ell, x_0, v_0, I, r, gamma), vel_1)
vel_2_f = sp.lambdify((t, g, m_1, m_2, ell, x_0, v_0, I, r, gamma), vel_2)

pos_1_array = pos_1_f(time, *parameters)
pos_2_array = pos_2_f(time, *parameters)

vel_1_array = vel_1_f(time, *parameters)
vel_2_array = vel_2_f(time, *parameters)

# cut off arrays when displacement is equal to the length of the rope
idx = np.argmax(pos_1_array > param_ell)
time = time[:idx]
pos_1_array = pos_1_array[:idx]
pos_2_array = pos_2_array[:idx]
vel_1_array = vel_1_array[:idx]
vel_2_array = vel_2_array[:idx]

np.savetxt(
    'atwood_machine.csv',
    np.transpose([time, pos_1_array, pos_2_array, vel_1_array, vel_2_array]),
    delimiter=',',
    header='time, pos_1, pos_2, vel_1, vel_2',
)
```

We will now animate the Atwood machine. We will use the `FuncAnimation` function from `matplotlib.animation` to animate the system.


```python
from matplotlib import patches
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))

ax.set_facecolor("k") # enable this to hide background
ax.get_xaxis().set_ticks([]) # enable this to hide x axis ticks
ax.get_yaxis().set_ticks([]) # enable this to hide y axis ticks

pulley = patches.Circle((0, 0), param_r, color="#66023c", fill=False, linewidth=2)
ax.add_patch(pulley)
ax.set_aspect("equal")

mass_1, = ax.plot([], [], "o", color="#66023c", markersize=param_m_1 * 5)
mass_2, = ax.plot([], [], "o", color="#66023c", markersize=param_m_2 * 5)

rope_1, = ax.plot([], [], color="#66023c", linewidth=2, linestyle="--")
rope_2, = ax.plot([], [], color="#66023c", linewidth=2, linestyle="--")

ax.set_xlim(-1.5 * (param_r + param_m_1 * 5), 1.5 * (param_r + param_m_2 * 5))
ax.set_ylim(-1.5 * param_ell, 1.5 * param_ell)

def animate(i):
    # have masses move up and down
    mass_1.set_data(-param_r, -pos_1_array[i])
    mass_2.set_data(param_r, -pos_2_array[i])
    
    # have ropes move up and down
    # make the dashes appear to move by shifting where they start
    rope_1.set_data([-param_r, -param_r], [0, -pos_1_array[i]])
 
    rope_2.set_data([param_r, param_r], [0, -pos_2_array[i]])

    # If the mass were to move up by 0.1m, I would want every dash on the corresponding rope
    # to be 0.1m above where they were previously. This is what the following code does.
    
    dash_length = param_ell / 6
    dash_shift = (i / 2) % 1
    
    # If the mass is going up, I want the dashes to go up as well. 
    # If the mass is going down, I want the dashes to go down as well.
    
    if vel_1_array[i] > 0:
        rope_1.set_dashes((0, dash_shift * dash_length, dash_length, (1 - dash_shift) * dash_length))
    else:
        rope_1.set_dashes((0, (1 - dash_shift) * dash_length, dash_length, dash_shift * dash_length))
        
    if vel_2_array[i] > 0:
        rope_2.set_dashes((0, dash_shift * dash_length, dash_length, (1 - dash_shift) * dash_length))
    else:
        rope_2.set_dashes((0, (1 - dash_shift) * dash_length, dash_length, dash_shift * dash_length))
        
   
ani = animation.FuncAnimation(fig, animate, frames=len(time), interval=50)
ani.save("Animations/atwood_machine.gif", writer="pillow", fps=25)
```

```python
# Kinetic energy
kinetic_f = sp.lambdify((m_1, m_2, ell, x_dot, I, r), T)

kinetic_array = kinetic_f(param_m_1, param_m_2, param_ell, vel_1_array, param_I, param_r)

# Potential energy
potential_f = sp.lambdify((g, m_1, m_2, ell, x), V)

potential_array = potential_f(param_g, param_m_1, param_m_2, param_ell, pos_1_array)
```

```python
plt.plot(time, kinetic_array)
plt.plot(time, potential_array)

plt.plot(time, kinetic_array + potential_array)

plt.title("Energy vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Energy (J)")
plt.legend(["Kinetic Energy", "Potential Energy", "Total Energy"], loc="upper right")
plt.grid()

plt.savefig("Figures/atwood_machine_energy.png", dpi=300)
plt.show()
```

```python
momentum_1 = m_1 * x_dot
momentum_2 = m_2 * x_dot

momentum_1_f = sp.lambdify((m_1, x_dot), momentum_1)
momentum_2_f = sp.lambdify((m_2, x_dot), momentum_2)

momentum_1_array = momentum_1_f(param_m_1, vel_1_array)
momentum_2_array = momentum_2_f(param_m_2, vel_2_array)

total_momentum_array = momentum_1_array + momentum_2_array
```
