# Solve Simultaneous Equations:

## $$4x^2 + 2y = 1$$
## $$2x + 3y^2 = 2$$

```python
# Create function 4x^2 + 2y - 1 = 0

x, y = sym.symbols("x y")

f1 = 4 * x**2 + 2 * y - 1

# Create function 2x + 3y^2 - 2 = 0

f2 = 2 * x + 3 * y**2 - 2

# Solve the simultaneous equations

sol = sym.solve([f1, f2], [x, y])

value_x, value_y = sol[0], sol[1]
```

```python
# Rearange f1 to get y

f1_for_y = sym.Eq(y, (1 - 4 * x**2) / 2)

# Get right hand side of equality and use to plot

sym.plotting.plot(f1_for_y.rhs)
```

```python
# Rearange set expression f2 to 0

f2_to_zero = sym.Eq(f2, 0)

# Rearange f2 to get y

solved = sym.solve(f2_to_zero, y)

f2_for_y = sym.Eq(y, solved[0]), sym.Eq(y, solved[1])

f2_for_y[0]
```

```python
f2_for_y[1]
```

```python
sym.plotting.plot(f2_for_y[0].rhs, f2_for_y[1].rhs)
```

```python
# Plotting both equations

p1 = sym.plotting.plot(f1_for_y.rhs, show=False)
p2 = sym.plotting.plot(f2_for_y[0].rhs, f2_for_y[1].rhs, show=False)

p2[0].line_color, p2[1].line_color = "red", "red"

p1.extend(p2)

p1.ylim = (-5, 5)
p1.xlim = (-5, 5)
p1.title = "$4x^2 + 2y = 1$" + "(blue) and " + " $2x + 3y^2 = 2$" + "(red)"

p1.show()
```

# Find the points of intersection
## f2 is split into 2 equations, so we check intersection for both

```python
intersections_1 = sym.solve([f1_for_y, f2_for_y[0]], [x, y], dict=True)
```

```python
intersections_2 = sym.solve([f1_for_y, f2_for_y[1]], [x, y], dict=True)
```

# Check if points are imaginary

```python
real_intersections = []
for i in intersections_1:
    if i[x].is_real and i[y].is_real:
        real_intersections.append(i)

for i in intersections_2:
    if i[x].is_real and i[y].is_real:
        real_intersections.append(i)

real_intersections
```

# Plot the points of intersection

```python
int_sect1 = sns.scatterplot(x=[real_intersections[0][x]], y=[real_intersections[0][y]], color="red").set_title("Intersection 1")
```

```python
int_sect2 = sns.scatterplot(x=[real_intersections[1][x]], y=[real_intersections[1][y]], color="red").set_title("Intersection 2")
```

# Create line from points of intersection

```python
# The formula for a line is y = mx + c

# m = (y2 - y1) / (x2 - x1)

m = (real_intersections[1][y] - real_intersections[0][y]) / (real_intersections[1][x] - real_intersections[0][x])

# c = y - mx

c = real_intersections[0][y] - m * real_intersections[0][x]

# Create equation for line

line = sym.Eq(y, m * x + c)
```

# Plot the line

```python
intersect_line_plot = sym.plotting.plot(line.rhs, show=False)
```

```python
intersect_line_plot.show()
```
